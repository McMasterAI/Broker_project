```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
X=np.asarray(stocks)
scale = StandardScaler()
X = scale.fit_transform(X)
X.shape
```

```python
n_clusters=30
cost=[]
for i in range(1,n_clusters):
    kmean= KMeans(i)
    kmean.fit(X)
    cost.append(kmean.inertia_)

plt.plot(range(1,30),cost, 'bx-')
```

```python
kmean= KMeans(7)
kmean.fit(X)
labels=kmean.labels_
clusters=pd.concat([stocks, pd.DataFrame(labels,index=stocks.index,columns=['clusters'])], axis=1)
clusters.head()
```

```python
for c in clusters:
    # print(c)

    grid= sns.FacetGrid(clusters, col='clusters')
    grid.map(plt.hist, c)

```

```python
pca = PCA(n_components =2)

X_PCA = pca.fit_transform(X)
X_PCA.shape
```

```python
labels_PCA = KMeans(n_clusters = 7).fit_predict(X_PCA)
# Visualizing the clustering
plt.scatter(X_PCA[:, 0], X_PCA[:, 1],
           c = labels_PCA, cmap =plt.cm.nipy_spectral)
plt.show()
```
